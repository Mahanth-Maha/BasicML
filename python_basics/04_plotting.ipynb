# IRIS DataSet - EDA

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

iris_dataset = load_iris()
iris_dataset
```

```python
X = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
y = pd.DataFrame(iris_dataset.target, columns=['species'])
X
y
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.2
X_train, X_test, y_train , t_test = train_test_split(X, y, test_size=test_size)
X_train.shape, X_test.shape, y_train.shape , t_test.shape
```

```python
for i, target_name in enumerate(iris_dataset['target_names']):
	plt.scatter(iris_feat[iris_target == i, 0], iris_feat[iris_target == i, 1], label=target_name)
plt.title('Iris dataset')
plt.xlabel(iris_dataset['feature_names'][0])
plt.ylabel(iris_dataset['feature_names'][1])
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X['sepal length (cm)'], X['sepal width (cm)'], X['petal length (cm)'], c=iris_target, cmap='viridis')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
plt.title('3D plot of Iris dataset')
plt.show()
```

```python
iris_df = pd.concat([X, y], axis=1)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))
plt.suptitle('Pair Plot of Iris Dataset', size=16)

for i in range(4):
    for j in range(4):
        if i == j:
            axes[i, j].hist(iris_df.iloc[:, i], bins=20, color='gray')
        else:
            for species in np.unique(iris_target):
                axes[i, j].scatter(
                    iris_df[iris_df['species'] == species].iloc[:, j],
                    iris_df[iris_df['species'] == species].iloc[:, i],
                    label=iris_dataset['target_names'][species],
                    s=7
                )
        if i == 3:
            axes[i, j].set_xlabel(iris_dataset['feature_names'][j])
        if j == 0:
            axes[i, j].set_ylabel(iris_dataset['feature_names'][i])

handles, labels = axes[0, 1].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')
plt.show()
```

```python

# Create a pair plot
import seaborn as sns
sns.set_style("whitegrid")
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
setosa = iris_feat[iris_target == 0]
spl_counts , spl_bins = np.histogram(setosa[:,0], bins = 10)
pdf_setosa_pl = spl_counts / np.sum(spl_counts)
cdf_setosa_pl = np.cumsum(pdf_setosa_pl)
plt.plot(pdf_setosa_pl)
plt.plot(cdf_setosa_pl)
```

```python
def mad(data):
    data = np.array(data)
    return np.median(np.abs(data - np.median(data)))

from statsmodels import robust

np.mean(setosa[:,0]), np.std(setosa[:,0])
np.median(setosa[:,0]) , mad(setosa[:,0]), robust.mad(setosa[:,0])
```

## f-string


```python
#### detour f - string 
a,b,c,d,e,f,g = 1e9, 1e-4, 1244,78**-2.5, 7.4213e-4, 10e4 , 7**8
# separate '_' or ',' s can be used
print(f'{a:,}') 
# algin text
print(f'{a:.10f}')
print(f'{b:>20}\t{c:>10}') # b takes 20 chars , c takes 10  RIGHT ALIGN
print(f'{b:<20}\t{c:<10}') # b takes 20 chars , c takes 10  LEFT  ALIGN (DEFAULT)
print(f'{b:^20}\t{c:^10}') # b takes 20 chars , c takes 10  CENTRE  ALIGN
# filling spaces with chars '_'
print(f'{b:_>20}\t{c:+>10}') # b takes 20 chars , c takes 10  RIGHT ALIGN
print(f'{b:*<20}\t{c:!<10}') # b takes 20 chars , c takes 10  LEFT  ALIGN (DEFAULT)
print(f'{b:%^20}\t{c:&^10}') # b takes 20 chars , c takes 10  CENTRE  ALIGN
```

```python
# can also do datatimes 
from datetime import datetime 
now = datetime.now()
print(f'{now}')
print(f'{now:%d__%m__%Y}')
print(f'{now:%D at %H:%M:%S}')
```

```python
# rounding of with f-string
print(f'{d} vs {d:.6f}')
print(f'{a} vs {a:20.10f}')
print(f'{f} vs {f:.2} vs {f:.5f}')

print(f'{a + b = }')
print(f'{c * e = }')
print(f'{e / f = :>50.10f}')
```

```python
movie_years = {}
for j in range(len(movies)):
    for i in re.compile(r'\d{4}').finditer(movies[j]):
        # print(f'movie {j} year : {i.group()}')
        if movie_years.get(j) == None:
            movie_years[j] = []
        movie_years[j].append(i.group())
movie_years
```

```python
movie_years = {}
for j in range(len(movies)):
    for i in re.compile(r'\(\d{4}\)').finditer(movies[j]):
        # print(f'movie {j} year : {i.group()}')
        if movie_years.get(j) == None:
            movie_years[j] = []
        movie_years[j].append(i.group().split('(')[1].split(')')[0])
movie_years
```

```python
questions_marks = re.compile(r'\(\?{4}/?I?I?\)')
data2 = questions_marks.sub( '' , data)
len(data), len(data2)
```

```python
def clean_years(x):
    with_is = re.compile(r'(\(\d{4})/I+\)')
    x = with_is.sub( r'\1)' , x)
    x = re.compile(r':::').sub( '' , x)
    x = re.compile(r'  ').sub( ' ' , x)
    return x.strip()

data2 = clean_years(data2)
movies = data2.split('\n')
len(movies)
```

```python
def get_year(x):
    year  = re.compile(r'\(\d{4}\)').findall(x)
    if len(year) > 0:
        return year[0].split('(')[1].split(')')[0]
    else:
        return None
```

```python
df = pd.DataFrame(movies, columns = ['Movies'])
df['Year'] = df['Movies'].apply(get_year)
df
```

```python
df[df['Year'].isnull()].shape[0]
```

```python
df['comedy'] = df['Movies'].apply(lambda x: len(re.compile(r'\bcomedy\b').findall(x)))
df['drama'] = df['Movies'].apply(lambda x: len(re.compile(r'\bdrama\b').findall(x)))
df['thriller'] = df['Movies'].apply(lambda x: len(re.compile(r'\bthriller\b').findall(x)))
df['romance'] = df['Movies'].apply(lambda x: len(re.compile(r'\bromance\b').findall(x)))

df
```

```python
df['Movies'] = df['Movies'].apply( lambda x : ' '.join(x.split(' ')[1:]))
df
```

```python
df.describe()
```

```python
df.rename(columns = {'Movies':'description'}, inplace = True)
df
```

```python
df['Movie'] = df['description'].apply(lambda x: x.split('(')[0].strip().replace('"','').replace("'",'').replace("",'').replace('  ',' '))
df
```

# Text Preprocessing 

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
cv = CountVectorizer( stop_words = 'english') 
cv.fit(df['description'])
cv_data = cv.transform(df['description'])
cv_data.toarray().shape
```

```python
countv_df = pd.DataFrame(cv_data.toarray())
countv_df.columns = cv.get_feature_names_out()
countv_df
```

```python
countv_df.sum().sort_values(ascending = False)
```

#### TF-IDF

```python
tfidf = TfidfVectorizer( stop_words = 'english')
tfidf.fit(df['description'])
Tfidf_data = tfidf.transform(df['description'])
Tfidf_data.toarray().shape
```

```python
Tfidf_df = pd.DataFrame(Tfidf_data.toarray())
Tfidf_df.shape
```

```python
Tfidf_df.columns = tfidf.get_feature_names_out()
Tfidf_df
```

```python
Tfidf_df.sum().sort_values(ascending = False)
```

```python
columns_to_drop = Tfidf_df.sum()[Tfidf_df.sum() < 1].index
Tfidf_df.drop(columns = columns_to_drop, inplace = True)
Tfidf_df.shape
```

```python
import numpy as np

def generate_spiral_data(points, classes):
    X = []
    y = []
    for class_number in range(classes):
        for point_number in range(points):
            k = 10
            r = point_number / points
            t = class_number * k + r * k + np.random.uniform(0, 0.2) 
            noise1, noise2 = np.random.uniform(-0.1, 0.1) , np.random.uniform(0.0, 0.05)
            X.append([r * np.sin(t) + noise1, r * np.cos(t) + noise2]) 
            y.append(class_number)
    return np.array(X), np.array(y)

X, y = generate_spiral_data(750, 3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Spiral Pattern")
plt.show()


# df = pd.DataFrame(X, columns=['x', 'y'])
# df['target'] = y
# df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N.csv', index=False) 

# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_X.npy', X)
# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_y.npy', y)

# dataset_counter += 1
# counter_2d += 1

df = pd.DataFrame(X, columns=['x', 'y'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_y.npy', y)

dataset_counter += 1
counter_2d += 1
```

```python
def star_pattern(n_samples=300, n_classes=3, n_spokes=5):
    X, y = [], []
    angles = np.linspace(0, 2 * np.pi, n_spokes, endpoint=False)
    for class_label in range(n_classes):
        for angle in angles:
            r = np.linspace(0.1, 1, n_samples // (n_classes * n_spokes))
            x1 = r * np.cos(angle) + np.random.normal(0, 0.05, len(r))
            x2 = r * np.sin(angle) + np.random.normal(0, 0.05, len(r)) # + np.random.normal(-0.5, 0.5)
            X.extend(np.c_[x1, x2])
            y.extend([class_label] * len(r))
    return np.array(X), np.array(y)
star_pattern_X, star_pattern_y = star_pattern()
plt.scatter(star_pattern_X[:, 0], star_pattern_X[:, 1], c=star_pattern_y, cmap='viridis')
plt.title("Star Pattern")
plt.show()
```

```python
def checkerboard_pattern(n_samples=300, n_classes=2, grid_size=5):
    X, y = [], []
    n_points = n_samples // (grid_size ** 2)
    for i in range(grid_size):
        for j in range(grid_size):
            x1 = np.random.uniform(i, i + 1, n_points)
            x2 = np.random.uniform(j, j + 1, n_points)
            X.extend(np.c_[x1, x2])
            y.extend([((i + j) % n_classes)] * n_points)
    return np.array(X), np.array(y)

X, y = checkerboard_pattern(700)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Checkerboard Pattern")
plt.show()

df = pd.DataFrame(X, columns=['x', 'y'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_y.npy', y)

dataset_counter += 1
counter_2d += 1
```

```python

def waves_pattern(n_samples=300, n_classes=3):
    X, y = [], []
    x_vals = np.linspace(0, 2 * np.pi, n_samples // n_classes)
    for class_label in range(n_classes):
        amplitude = class_label + 1
        phase = class_label * np.pi / 4
        x1 = x_vals + np.random.normal(0, 0.1, len(x_vals))
        x2 = amplitude * np.sin(x_vals + phase) + np.random.normal(0, 0.1, len(x_vals))
        X.extend(np.c_[x1, x2])
        y.extend([class_label] * len(x_vals))
    return np.array(X), np.array(y)

X, y = waves_pattern()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Waves Pattern")
plt.show()

# df = pd.DataFrame(X, columns=['x', 'y'])
# df['target'] = y
# df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N.csv', index=False) 

# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_X.npy', X)
# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_y.npy', y)

# dataset_counter += 1
# counter_2d += 1
```

```python
def concentric_rings(n_samples=750, n_classes=3):
    X, y = [], []
    for class_label in range(n_classes):
        radius = (class_label + 1 ) * 1.1
        angles = np.linspace(0, 2 * np.pi, n_samples // n_classes)
        x1 = radius * np.cos(angles) + np.random.normal(0, 0.3, len(angles))
        x2 = radius * np.sin(angles) + np.random.normal(0, 0.2, len(angles))
        X.extend(np.c_[x1, x2])
        y.extend([class_label] * len(angles))
    return np.array(X), np.array(y)

X, y = concentric_rings()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Concentric Rings")
plt.show()

# df = pd.DataFrame(X, columns=['x', 'y'])
# df['target'] = y
# df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N.csv', index=False) 

# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_X.npy', X)
# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_y.npy', y)

# dataset_counter += 1
# counter_2d += 1

df = pd.DataFrame(X, columns=['x', 'y'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_y.npy', y)

dataset_counter += 1
counter_2d += 1
```

```python
def xor_pattern(n_samples=300, gap=0.2):
    n_per_quad = n_samples // 4
    X, y = [], []
    for i, (x_shift, y_shift) in enumerate([(0, 0), (0, 1), (1, 0), (1, 1)]):
        x = np.random.uniform(0, 1, n_per_quad) + x_shift
        y_coords = np.random.uniform(0, 1, n_per_quad) + y_shift
        X.extend(np.c_[x, y_coords])
        y.extend([0 if (x_shift + y_shift) % 2 == 0 else 1] * n_per_quad)
    X = np.array(X) - 0.5  # Center the pattern
    return X, np.array(y)

X, y = xor_pattern()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("XOR Pattern")
plt.show()


df = pd.DataFrame(X, columns=['x', 'y'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_y.npy', y)

dataset_counter += 1
counter_2d += 1
```

```python
def gaussian_clusters_2d(n_samples=300, n_classes=3):
    X, y = [], []
    for class_label in range(n_classes):
        mean = np.random.uniform(-5, 5, 2)
        cov = np.diag(np.random.uniform(0.5, 1.5, 2))
        points = np.random.multivariate_normal(mean, cov, n_samples // n_classes)
        X.extend(points)
        y.extend([class_label] * (n_samples // n_classes))
    return np.array(X), np.array(y)

X, y = gaussian_clusters_2d()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Gaussian Clusters")


df = pd.DataFrame(X, columns=['x', 'y'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_y.npy', y)

dataset_counter += 1
counter_2d += 1
```

```python
def circular_waves(n_samples=750, n_classes=3, noise=0.4):
    X, y = [], []
    points_per_wave = n_samples // n_classes
    for class_label in range(n_classes):
        radius = (class_label + 1) * 2
        theta = np.linspace(0, 2 * np.pi, points_per_wave)
        r = radius + np.sin(4 * theta) + np.random.normal(0, noise, points_per_wave)
        x = r * np.cos(theta)
        y_coords = r * np.sin(theta)
        X.extend(np.c_[x, y_coords])
        y.extend([class_label] * points_per_wave)
    return np.array(X), np.array(y)

circular_X, circular_y = circular_waves()
plt.scatter(circular_X[:, 0], circular_X[:, 1], c=circular_y, cmap='viridis')
plt.title("Circular Waves")
plt.show()

# df = pd.DataFrame(X, columns=['x', 'y'])
# df['target'] = y
# df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N.csv', index=False) 

# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_X.npy', X)
# np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_2N_y.npy', y)

# dataset_counter += 1
# counter_2d += 1
df = pd.DataFrame(X, columns=['x', 'y'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_2D_{counter_2d}_3N_y.npy', y)

dataset_counter += 1
counter_2d += 1
```

```python
import numpy as np

def concentric_spheres3D(n_samples=300, n_classes=3):
    X, y = [], []
    for class_label in range(n_classes):
        radius = (class_label + 1) * 2
        points = n_samples // n_classes
        phi = np.random.uniform(0, np.pi, points)  # Inclination angle
        theta = np.random.uniform(0, 2 * np.pi, points)  # Azimuthal angle
        x = radius * np.sin(phi) * np.cos(theta) + np.random.normal(0, 0.1, points)
        y_coords = radius * np.sin(phi) * np.sin(theta) + np.random.normal(0, 0.1, points)
        z = radius * np.cos(phi) + np.random.normal(0, 0.1, points)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points)
    return np.array(X), np.array(y)

X, y = concentric_spheres3D(500, 2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
plt.title("Concentric Spheres")
plt.show()

df = pd.DataFrame(X, columns=['x', 'y' ,'z'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_y.npy', y)

dataset_counter += 1
counter_3d += 1

```

```python
def helix_pattern(n_samples=300, n_classes=2, n_turns=3):
    X, y = [], []
    for class_label in range(n_classes):
        points = n_samples // n_classes
        t = np.linspace(0, n_turns * 2 * np.pi, points)  # Angle parameter
        z = np.linspace(0, 10, points) + class_label * 5  # Vertical offset
        x = np.cos(t) + np.random.normal(0, 0.1, points)
        y_coords = np.sin(t) + np.random.normal(0, 0.1, points)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points)
    return np.array(X), np.array(y)

X, y = helix_pattern()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
plt.title("Helix Pattern")
plt.show()

df = pd.DataFrame(X, columns=['x', 'y' ,'z'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_y.npy', y)

dataset_counter += 1
counter_3d += 1
```

```python
def checkerboard_3d(n_samples=750, grid_size=5, n_classes=2):
    X, y = [], []
    n_points = n_samples // (grid_size ** 3)
    for i in range(grid_size):
        for j in range(grid_size):
            for k in range(grid_size):
                x = np.random.uniform(i, i + 1, n_points)
                y_coords = np.random.uniform(j, j + 1, n_points)
                z = np.random.uniform(k, k + 1, n_points)
                X.extend(np.c_[x, y_coords, z])
                label = (i + j + k) % n_classes
                y.extend([label] * n_points)
    return np.array(X), np.array(y)

checkerboard_X, checkerboard_y = checkerboard_3d()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(checkerboard_X[:, 0], checkerboard_X[:, 1], checkerboard_X[:, 2], c=checkerboard_y, cmap='viridis')
plt.title("Checkerboard 3D")
plt.show()
```

```python
def waves_3d_pattern(n_samples=300, n_classes=3):
    X, y = [], []
    for class_label in range(n_classes):
        points = n_samples // n_classes
        x = np.linspace(0, 10, points)
        y_coords = np.linspace(0, 10, points)
        z = np.sin(x + class_label) + np.cos(y_coords - class_label) + np.random.normal(0, 0.1, points)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points)
    return np.array(X), np.array(y)

waves_X, waves_y = waves_3d_pattern()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(waves_X[:, 0], waves_X[:, 1], waves_X[:, 2], c=waves_y, cmap='viridis')
plt.title("Waves 3D Pattern")
plt.show()
```

```python
def spherical_shells(n_samples=300, n_classes=3):
    X, y = [], []
    for class_label in range(n_classes):
        radius = (class_label + 1) * 3
        points = n_samples // n_classes
        phi = np.random.uniform(0, np.pi, points)
        theta = np.random.uniform(0, 2 * np.pi, points)
        x = radius * np.sin(phi) * np.cos(theta) + np.random.normal(0, 0.2, points)
        y_coords = radius * np.sin(phi) * np.sin(theta) + np.random.normal(0, 0.2, points)
        z = radius * np.cos(phi) + np.random.normal(0, 0.2, points)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points)
    return np.array(X), np.array(y)

spherical_X, spherical_y = spherical_shells()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(spherical_X[:, 0], spherical_X[:, 1], spherical_X[:, 2], c=spherical_y, cmap='viridis')
plt.title("Spherical Shells")
plt.show()
```

```python
def clustered_planes(n_samples=500, n_classes=2, plane_gap=2):
    X, y = [], []
    for class_label in range(n_classes):
        z = class_label * plane_gap + np.random.normal(0, 0.2, n_samples // n_classes)
        x = np.random.uniform(-10, 10, n_samples // n_classes)
        y_coords = np.random.uniform(-10, 10, n_samples // n_classes)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * (n_samples // n_classes))
    return np.array(X), np.array(y)

X, y = clustered_planes(500,2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
plt.title("Clustered Planes")
plt.show()


df = pd.DataFrame(X, columns=['x', 'y' ,'z'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_y.npy', y)

dataset_counter += 1
counter_3d += 1

```

```python
def gaussian_clusters_3d(n_samples=300, n_classes=2):
    X, y = [], []
    for class_label in range(n_classes):
        mean = np.random.uniform(-10, 10, 3)
        cov = np.diag(np.random.uniform(0.5, 2, 3))
        points = np.random.multivariate_normal(mean, cov, n_samples // n_classes)
        X.extend(points)
        y.extend([class_label] * (n_samples // n_classes))
    return np.array(X), np.array(y)

X, y = gaussian_clusters_3d(500, 2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
plt.title("Gaussian Clusters 3D")
plt.show()

df = pd.DataFrame(X, columns=['x', 'y' ,'z'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_y.npy', y)

dataset_counter += 1
counter_3d += 1
```

```python
def spiral_stairs(n_samples=500, n_classes=2, height=50, turns=4):
    X, y = [], []
    points_per_class = n_samples // n_classes
    for class_label in range(n_classes):
        t = np.linspace(0, 2 * np.pi * turns, points_per_class)
        z = np.linspace(class_label * height / n_classes, (class_label + 1) * height / n_classes, points_per_class)
        x = np.sin(t) + np.random.normal(0, 0.1, points_per_class)
        y_coords = np.cos(t) + np.random.normal(0, 0.1, points_per_class)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points_per_class)
    return np.array(X), np.array(y)

spiral_X, spiral_y = spiral_stairs()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(spiral_X[:, 0], spiral_X[:, 1], spiral_X[:, 2], c=spiral_y, cmap='viridis')
plt.title("Spiral Stairs")
plt.show()

df = pd.DataFrame(X, columns=['x', 'y' ,'z'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_y.npy', y)

dataset_counter += 1
counter_3d += 1
```

```python
def spheres_in_cube(n_samples=1000, n_classes=2, cube_size=10):
    X, y = [], []
    for class_label in range(n_classes):
        radius = (class_label + 1) * (cube_size / (2 * n_classes))
        center = np.random.uniform(-cube_size / 2, cube_size / 2, 3)
        points = n_samples // n_classes
        phi = np.random.uniform(0, np.pi, points)
        theta = np.random.uniform(0, 2 * np.pi, points)
        x = center[0] + radius * np.sin(phi) * np.cos(theta) + np.random.normal(0, 0.1, points)
        y_coords = center[1] + radius * np.sin(phi) * np.sin(theta) + np.random.normal(0, 0.1, points)
        z = center[2] + radius * np.cos(phi) + np.random.normal(0, 0.1, points)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points)
    return np.array(X), np.array(y)

X, y = spheres_in_cube(n_classes=2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
plt.title("Spheres in Cube")
plt.show()

df = pd.DataFrame(X, columns=['x', 'y' ,'z'])
df['target'] = y
df.to_csv(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N.csv', index=False) 

np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_X.npy', X)
np.save(f'{data_folder}{"0"*(2 - len(str(dataset_counter)))}{dataset_counter}_3D_{counter_3d}_2N_y.npy', y)

dataset_counter += 1
counter_3d += 1


```

```python
def zigzag_layers(n_samples=300, n_classes=2, amplitude=3, gap=2):
    X, y = [], []
    points_per_layer = n_samples // n_classes
    for class_label in range(n_classes):
        z = class_label * gap + np.random.normal(0, 0.2, points_per_layer)
        x = np.linspace(-10, 10, points_per_layer)
        y_coords = amplitude * np.sin(0.5 * x) + np.random.normal(0, 0.2, points_per_layer)
        X.extend(np.c_[x, y_coords, z])
        y.extend([class_label] * points_per_layer)
    return np.array(X), np.array(y)

zigzag_X, zigzag_y = zigzag_layers()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(zigzag_X[:, 0], zigzag_X[:, 1], zigzag_X[:, 2], c=zigzag_y, cmap='viridis')
plt.title("Zigzag Layers")
plt.show()
```

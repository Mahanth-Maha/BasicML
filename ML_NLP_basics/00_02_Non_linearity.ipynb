# sigmoid function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def plot_function(func = sigmoid, text = 'sigmoid' , x_range = (-10, 10), y_range = (0, 1)):
    x = np.linspace(*x_range, 100)
    y = func(x)
    _ = plt.plot(x, y , label = text)
    _ = plt.title(f'{text} Function')
    _ = plt.xlabel('x')
    _ = plt.ylabel(f'{text}(x)')
    _ = plt.xlim(x_range)
    _ = plt.ylim(y_range)
    _ = plt.grid(True)
    _ = plt.legend()
    _ = plt.show()

plot_function()
```

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

plot_function(func = sigmoid_derivative, text = 'sigmoid_derivative')
```

```python
def sigmoid_move(x , move_right = 0 , move_top = 5):
    return sigmoid(x - move_right) + move_top
plot_function(func = sigmoid_move, text = 'sigmoid_moved' , x_range = (-2, 2), y_range = (0, 10))
```

```python
def sigmoid_(x ,a=5,b=2,c=2,d=2):
    return a * sigmoid(b * (x - c)) + d

plot_function(func = sigmoid_, text = 'sigmoid_', x_range = (-10, 10), y_range = (0, 10))
```
